```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
import numpy as np

vocab_size = 88000

maxlen = 250


# Load the data, keeping only the top `vocab_size` words.
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=vocab_size)

print(f"Training sequences: {len(x_train)}")
print(f"Test sequences: {len(x_test)}")
```

```python
# Pre-padding is applied by default (adding zeros at the start).
x_train = sequence.pad_sequences(x_train, maxlen=maxlen)
x_test = sequence.pad_sequences(x_test, maxlen=maxlen)

print("Padded Training data shape:")
print(x_train.shape)
print("Padded Test data shape:")
print(x_test.shape)
```

```python
embedding_dim = 32
rnn_units = 32

model = Sequential()
# Layer 1: Embedding Layer
model.add(Embedding(vocab_size, embedding_dim ))
# Layer 2: SimpleRNN Layer
model.add(SimpleRNN(rnn_units))
# Layer 3: Output Dense Layer with Sigmoid activation for probability output
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
vocab_size = 88000
maxlen = 250
embedding_dim = 32
rnn_units = 32

model = Sequential()

model.add(Embedding(vocab_size + 1, embedding_dim, input_length=maxlen))
model.add(SimpleRNN(rnn_units))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
model.compile(
    optimizer='rmsprop',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    x_train,
    y_train,
    epochs=40,
    validation_split=0.25
)
```

```python
loss, accuracy = model.evaluate(x_test, y_test, verbose=0)

print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")


sample_review = x_test[0]

prediction = model.predict(np.expand_dims(sample_review, axis=0))


print(f"\nPrediction Probability (0=Negative, 1=Positive): {prediction[0][0]:.4f}")

# Convert the probability to a binary label (using a 0.5 threshold)
predicted_label = 1 if prediction[0][0] >= 0.5 else 0

print(f"Actual Label (y_test[0]): {y_test[0]}")
print(f"Predicted Label: {predicted_label}")
```

```python
import gradio as gr
from tensorflow.keras.datasets import imdb
import numpy as np

# --- 1. Get Word Index Mapping ---
# The IMDb dataset uses 1-based indexing, and reserves indices 0-3 for special tokens.

word_index = imdb.get_word_index()

# Create a reverse index map for human-readable tokens
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])

# Add special tokens as per Keras IMDb convention:
# 0: Padding, 1: Start of sequence, 2: Unknown word, 3: Unused
# The index in the data is actually shifted by 3 from the raw index.
def decode_review(text_indices):
    return ' '.join([reverse_word_index.get(i - 3, '?') for i in text_indices])


# --- 2. Main Prediction Function ---

def predict_sentiment(raw_text):
    """
    Converts raw text into a sequence of indices, pads it, and predicts sentiment.
    """
    # Tokenize the input text using the word index
    tokens = []

    # 1. Start token (index 1)
    tokens.append(1)

    # 2. Convert words to indices, shifting them by +3 to account for special tokens.
    # Words not in the vocab (key not found) are replaced by the unknown token (index 2).
    for word in raw_text.lower().split():
        index = word_index.get(word)
        if index is not None and index < vocab_size:
            # We use index + 3 because Keras IMDb reserves indices 0, 1, 2, 3
            tokens.append(index + 3)
        else:
            # Unknown word index (index 2)
            tokens.append(2)

    # 3. Apply padding and truncation (Matching maxlen=250)
    # The sequence.pad_sequences function is designed to handle this.
    # Since we are predicting a single review, we wrap the tokens in a list.
    padded_sequence = sequence.pad_sequences([tokens], maxlen=maxlen)

    # 4. Predict using the model
    # The model expects a batch of inputs, even for a single review.
    prediction = model.predict(padded_sequence, verbose=0)
    probability = prediction[0][0]

    # 5. Format Output
    if probability >= 0.5:
        sentiment = "POSITIVE Review ğŸ˜Š"
    else:
        sentiment = "NEGATIVE Review ğŸ˜ "

    result_text = f"Sentiment: {sentiment}\n"
    result_text += f"Probability of Positive: {probability:.4f}"

    return result_text
```

```python
iface = gr.Interface(
    fn=predict_sentiment,
    inputs=gr.Textbox(lines=5, label="Enter Movie Review (Text)"),
    outputs=gr.Textbox(label="Prediction Result"),
    title="Simple RNN IMDb Movie Review Sentiment Classifier",
    description="Enter a movie review to classify its sentiment (Positive/Negative) using a Keras Simple RNN model. The model was trained on the IMDb dataset.",
    examples=[
        ["This film was just brilliant. Casting, location, scenery, story, direction, everyone was really suited to the part they played."], # Example 1: Positive
        ["Big hair, big budgets, bad music, and cheap safety pins. These are the words to best describe this terrible movie."] # Example 2: Negative
    ]
)

# Launch the interface
iface.launch()
```
